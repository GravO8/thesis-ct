# Runs analysis

```python
import pandas as pd

model = "baseline-3DCNN-resnet18_gap_3D"
run1 = pd.read_csv(f"../../runs/systematic/baseline/{model}/{model}-run1/predictions.csv", sep = ";")
run2 = pd.read_csv(f"../../runs/systematic/baseline/{model}/{model}-run2/predictions.csv", sep = ";")
run3 = pd.read_csv(f"../../runs/systematic/baseline/{model}/{model}-run3/predictions.csv", sep = ";")

dataset = pd.read_csv("../../data/gravo/dataset.csv")
labels = []

i = 0
for id_ in run1["patient_id"].values:
    labels.append( dataset[dataset["patient_id"] == id_]["binary_rankin"].values[0] )
    assert run2["patient_id"].values[i] == id_
    assert run3["patient_id"].values[i] == id_
    i += 1

sapo = {}
sapo["ids"] = run1["patient_id"].values
sapo["run1"] = run1["y_prob"].values
sapo["run2"] = run2["y_prob"].values
sapo["run3"] = run3["y_prob"].values
sapo["y_trye"] = labels
sapo = pd.DataFrame(sapo)
sapo
```
